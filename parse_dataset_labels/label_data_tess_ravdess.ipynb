```python
import os

import random
from shutil import copyfile
import numpy
import csv
from PIL import Image
```

```python
# pic = path to picture
# original_dir = the directory this image is stored in
# emotion_ = which emotion it is to match the directory
# global_emotion_dir = the directory we wish to save the image in
# ttv = train, test, or validation folder
def copy_files(pic, orignal_dir, emotion_, global_emotion_dir, ttv, file_name=None):
    file_ = os.path.join(orignal_dir, pic)
    ttv_dir = os.path.join(global_emotion_dir, ttv)
    emotion_dir = os.path.join(ttv_dir, emotion_)
    if file_name==None:
        dest_ = os.path.join(emotion_dir, pic)
    else:
        dest_ = os.path.join(emotion_dir, file_name)
    if os.path.getsize(file_) != 0:
        copyfile(file_, dest_)
```

```python
tess_r = "./TESS_and_RAVDESS/"
global_emo_directory = "./emotions_5_sound/"
```

```python
emotions = os.listdir(tess_r)

for emotion in emotions:
    dir_ = os.path.join(tess_r, emotion)  
    if os.path.isdir(dir_):
        files = os.listdir(dir_)

        if emotion == "Angry":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "anger_disgust", global_emo_directory, train_test_validate)

        elif emotion == "Disgusted":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "anger_disgust", global_emo_directory, train_test_validate)

        elif emotion == "Fearful":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "surprise_fear", global_emo_directory, train_test_validate)

        elif emotion == "Happy":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "joy", global_emo_directory, train_test_validate)
                
        elif emotion == "Neutral":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "neutral", global_emo_directory, train_test_validate)

        elif emotion == "Sad":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "sadness", global_emo_directory, train_test_validate)

        elif emotion == "Suprised":
            for file_ in files:
                randint = random.random()
                train_test_validate = None
                if randint < 0.8:
                    train_test_validate = "train"
                elif randint >= 0.8 and randint < 0.9:
                    train_test_validate = "validation"
                else:       
                    train_test_validate = "test"
                copy_files(file_, dir_, "surprise_fear", global_emo_directory, train_test_validate)

        else:
            continue
    else:
        continue
```
