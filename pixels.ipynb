```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import csv
import random
import tensorflow as tf
```

(0=Angry, 1=Disgust, 2=Fear, 3=Happy, 4=Sad, 5=Surprise, 6=Neutral).

```python
labels = []
images = []
with open("fer2013.csv") as rf:
    next(rf)
    rows = csv.reader(rf, delimiter=",")
    count=0
    for row in rows:
        labels.append(row[0])
        pixels = row[1]
        pixels = pixels.split(" ")
        images.append(np.array_split(pixels, 48))
        count +=1
        if count>2:
            break
c = list(zip(labels, images))
random.shuffle(c)
labels, images = zip(*c)
xyz = np.array(images).astype(float)

x = xyz[2]
x = np.expand_dims(x, 2)
print(x.shape)
nn = tf.image.resize(x,(256,256))
x = np.squeeze(nn, axis=2)
# zeros = np.ones((1,1,3))
# fi = image*zeros
print(labels[2])
plt.imshow(x)




        
```
